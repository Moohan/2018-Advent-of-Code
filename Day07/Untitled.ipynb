```python
def switchPos(indexable, a, b):
    indexable[a], indexable[b] = indexable[b], indexable[a]
    return indexable
```

```python
import itertools

pairs = [list(pair) for pair in p.findall(input)]
tasks = sorted(list(set().union(*pairs)))
c, g, u, x = tasks.index("C"), tasks.index("G"), tasks.index("U"), tasks.index("X")
tasks = switchPos(tasks, 0, c)
tasks = switchPos(tasks, 1, g)
tasks = switchPos(tasks, 2, u)
tasks = switchPos(tasks, 3, x)

tasks
```

```python
def orderTasks(tasks, orders):
    change = False
    for order in orders:
        #print(order)
        if tasks.index(order[0]) > tasks.index(order[1]):
            #print(tasks)
            i = tasks.index(order[0])
            tasks[i], tasks[i-1] = tasks[i-1], tasks[i]
            #print(tasks)
            change = True
    if change:
        orderTasks(tasks, orders)
        
    return tasks
```

```python
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return itertools.chain.from_iterable(itertools.combinations(s, r) for r in range(len(s)+1) if r > 0)

print(powerset(pairs))
```
